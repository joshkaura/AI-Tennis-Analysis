# Get Dataset

```python

from roboflow import Roboflow
import os
from dotenv import load_dotenv

load_dotenv()
ROBOFLOW_API_KEY = os.environ["ROBOFLOW_API_KEY"]  # raises KeyError if missing

rf = Roboflow(api_key=ROBOFLOW_API_KEY)
project = rf.workspace("viren-dhanwani").project("tennis-ball-detection")
version = project.version(6)
dataset = version.download("yolov11")
                
```

```python
import shutil

shutil.move('tennis-ball-detection-6/train', 
            'tennis-ball-detection/tennis-ball-detection-6/train')

shutil.move('tennis-ball-detection-6/test', 
            'tennis-ball-detection/tennis-ball-detection-6/test')

shutil.move('tennis-ball-detection-6/valid', 
            'tennis-ball-detection/tennis-ball-detection-6/valid')
```

```python
from ultralytics import YOLO
from pathlib import Path

# Point this to the data.yaml for your exported dataset
DATA_YAML = "tennis-ball-detection-6/data.yaml"

# Choose a starting checkpoint (nano is fastest; x is heavier)
model = YOLO("yolo11n.pt")   # or "yolo11s.pt", "yolo11m.pt", "yolo11x.pt"

print("Starting Training")

results = model.train(
    data=DATA_YAML,
    imgsz=640,
    epochs=100,
    batch=16,          # lower if you hit OOM (e.g., 8 or 4)
    device="cpu",      # Mac Apple Silicon; use 0 for NVIDIA GPU; or "cpu"
    workers=4,
    project="runs",
    name="tennis_ball_yolo11",
    exist_ok=True
)

print("Done. Best weights:", Path(results.save_dir) / "weights" / "best.pt")
```
